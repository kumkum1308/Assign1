```python
#importing important libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

plt.style.use('ggplot')
sns.set_palette('viridis')
```

```python
#loading dataset
df = pd.read_csv('Economic Indicators And Inflation.csv')
df.head()
```

```python
# Check column names
df.columns
```

```python
#data cleaning
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('_(%)', '').str.replace('_(in_billion_usd)', '')
df.info()
df.isnull().sum()
df.describe()
```

```python
#. Univariate Analysis  
```

```python
# Distribution of Inflation Rate
plt.figure(figsize=(8, 5))
sns.histplot(df['inflation_rate'], bins=20, kde=True, color='blue')
plt.title('Distribution of Inflation Rate')
plt.xlabel('Inflation Rate (%)')
plt.show()
```

```python
# Count of Countries
plt.figure(figsize=(8, 5))
sns.countplot(x='country', data=df)
plt.title('Count of Records by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Multivariate Analysis  
```

```python
# Correlation Heatmap
numerical_df = df.select_dtypes(include='number')

plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Average Inflation Rate by Country
plt.figure(figsize=(8, 5))
sns.barplot(x='country', y='inflation_rate', data=df)
plt.title('Average Inflation Rate by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# GDP vs Inflation Rate
plt.figure(figsize=(8, 5))
sns.scatterplot(x='gdp', y='inflation_rate', data=df)
plt.title('GDP vs Inflation Rate')
plt.show()
```

```python
# Unemployment Rate vs Economic Growth
plt.figure(figsize=(8, 5))
sns.scatterplot(x='unemployment_rate', y='economic_growth', data=df)
plt.title('Unemployment Rate vs Economic Growth')
plt.show()
```

```python
#. Calculating Statistical Measures  
inflation_mean = df['inflation_rate'].mean()
inflation_median = df['inflation_rate'].median()
inflation_mode = df['inflation_rate'].mode()[0]

print(f"Mean Inflation Rate: {inflation_mean}")
print(f"Median Inflation Rate: {inflation_median}")
print(f"Mode Inflation Rate: {inflation_mode}")
```

```python
#inflation rate by country
plt.figure(figsize=(10, 6))
sns.boxplot(x='country', y='inflation_rate', data=df, palette='viridis')
plt.title('Inflation Rate by Country')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Inflation Rate (%)')
plt.show()
```

```python
# Pair Plot 
sns.pairplot(numerical_df)
plt.show()
```

## 8. Conclusions  
- Inflation rate varies significantly across countries and years.  
- GDP and inflation rate may show an inverse or direct relationship depending on economic conditions.  
- Countries with higher unemployment rates tend to show different economic growth patterns.  
- There are visible patterns between economic indicators like GDP, unemployment, and inflation.
